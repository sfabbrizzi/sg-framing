```python
import pandas as pd
import numpy as np

from pathlib import Path

from IPython.display import display, Image
from pigeon import annotate
```

```python
ROOT: Path = Path("../")
RESULTS: Path = ROOT / "results"
DATA: Path = ROOT / "data"
VISUAL_GENOME: Path = DATA / "visual_genome"
IMAGES: Path = VISUAL_GENOME / "images" / "VG_100K"
```

Load dataframe

```python
df: pd.DataFrame = pd.read_csv(RESULTS / "people_data.csv")
df.head()
```

```python
no_added: int = len(df[(df.used_people_detector == 1) & (df.contains_person == 1)])
total_no: int = len(df[df.used_people_detector == 1])
print(f"Number of images where people detector added a person: {no_added} out of {total_no} ({no_added/total_no:.2%})")
```

Select sample of images labelled positively by the model

```python
sample_pos: pd.DataFrame = df[(df.used_people_detector == 1) & (df.contains_person == 1)].sample(500, random_state=732)
sample_neg: pd.DataFrame = df[(df.used_people_detector == 1) & (df.contains_person == 0)].sample(500, random_state=732)
```

Images labelled as positive

```python
images = [IMAGES / f"{image_id}.jpg" for image_id in sample_pos.image_id]

def display_image(filename: Path) -> None:
    print(f"Displaying image: {filename}")
    display(Image(filename))


annotations_ = annotate(
    images,
    options=["True positive", "False Positive"],
    display_fn=display_image
)
```

```python
FDR: int = (np.array([ 1 if annotation[1] == "True positive" else 0 for annotation in annotations_ ]) == 0).sum() / 500
print(f"False Discovery Rate (FDR) on sample of 500 images: {FDR:.2%}")
```
